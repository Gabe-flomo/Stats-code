```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp
pd.set_option('display.max_columns', 25)
import re

%matplotlib inline
```

```python
df = pd.read_csv('tb.csv')
```

```python
df.head()
```

```python
# first step is to rename the columns into something thats more understandable
columns = df.columns
new = list()

for column in columns:
    print(column)
    # get the male and female columns only
    if column.startswith('m') or column.startswith('f'):
        # split the sex from the age
        sex = column[0]
        age = column[1:]
        # determine the age range
        if len(age)%2 == 0 and age[-1] != 'u':
            # find the middle of the string
            half = len(age)//2
            # insert a dash in the middle of the string
            age_range = age[:half] + '-' + age[half:] if int(age[:half]) < int(age[half:]) else age
        elif len(age)%2 != 0 and age[-1] != 'u':
            age_range = age[0] + '-' + age[1:]
        else:
            age_range = age
            
        column = sex + '_' + age_range
        new.append(column)
    else:
        new.append(column)
```

```python
df.columns = new
df.head().T
```

```python
df.drop(['m_0-4','m_5-14','m_0-14','m_u','f_0-4','f_5-14','f_0-14','f_u'],axis = 1, inplace = True)
df.head().T
```

```python
sexes = [sex for sex in df.columns if sex.startswith('m') or sex.startswith('f')]
melted = df.melt(['country','year'], sexes, 'sex','cases')
```

```python
melted
```

```python
melted['age'] = melted['sex'].apply(lambda x: x[2:])
melted['sex'] = melted['sex'].apply(lambda x: x[0])
melted
```

```python
melted = melted.dropna()
```

```python
final = melted[['country','year','age','sex','cases']]
final.sort_values(['country','year','age'], inplace = True)
final
```

```python
final = final.reset_index().drop('index', axis = 1)
```
