Its hypothesized that women will have more children if it is less likely that the childdren will survive. So we can plot the babies per woman on the x axis and the percentage of children surviving until age 5 on the y axis.

```python
# to filter by year
df[df.year == 1965].plot.scatter('babies_per_woman','age5_surviving',title = 'Babies per woman in 1965')
```

```python
df.plot.scatter('babies_per_woman','age5_surviving')
```

```python
def plotyear(year):
    # get the data from the specified year
    data = df[df.year == 1965]
    # make the size of the circles bigger based on the size of the population
    area = 5e-6 * data.population
    # make different regions different colors
    colors = data.region.map({'Africa':'#e5989b','Europe':'#b5838d','America':'#6d6875','Asia':'#ffcdb2'})
    
    
    data.plot.scatter('babies_per_woman','age5_surviving',title = 'Babies per woman in 1965',
                      s=area,c=colors, figsize = (10,8), linewidths=1,edgecolor='k')
    
```

```python
plotyear(1965)
```

# Interactive Plotting

```python
# using Ipywidgets
interact(plotyear,year = widgets.IntSlider(min = 1950, max = 2015, step = 1, value = 1965))
```
